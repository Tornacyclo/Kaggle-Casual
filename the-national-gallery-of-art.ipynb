# Display

```python
import pandas as pd



# Read the CSV file
df = pd.read_csv('/kaggle/input/the-national-gallery-of-art-open-data-program/opendata-main/data/objects.csv')
```

```python
%%capture head
# Display the first 5 rows of the dataframe
print(df.head())
```

This code uses the pandas library to read the CSV file and store it in a dataframe. The pd.read_csv() function takes the file path as an argument. The df.head() function is then used to display the first five rows of the dataframe. If you want to display all rows, you can use print(df) instead.

# Analysis

In order to analyze the content of the CSV file, we can use various pandas functions to explore the data. Some common techniques include:

- Checking the shape of the dataframe using df.shape, which returns the number of rows and columns in the dataframe.
- Using df.info() to get a summary of the dataframe, including the number of non-null values for each column.
- Using df.describe() to get statistical information about the numerical columns in the dataframe.
- Using df.columns to get a list of the column names in the dataframe.
- Using df['column_name'].value_counts() to get the frequency of each unique value in a specific column.

```python
%%capture shape
print(df.shape)
```

```python
%%capture info
print(df.info())
```

```python
%%capture describe
print(df.describe())
```

```python
print(df.columns)
```

```python
print(df['classification'].value_counts())
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns



plt.rcParams["figure.figsize"] = [32, 18]
plt.rcParams["figure.dpi"] = 300



# Histogram of the beginyear column
sns.histplot(data = df, x = 'beginyear')
plt.show()
```

```python
# Bar plot of the classification column
sns.countplot(data = df, x = 'classification')
plt.show()
```

```python
# Scatter plot of beginyear and endyear columns
sns.scatterplot(data = df, x = 'beginyear', y = 'endyear')
plt.show()
```
