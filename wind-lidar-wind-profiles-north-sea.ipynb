```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt



plt.rcParams["figure.figsize"] = [32, 32]
plt.rcParams["figure.dpi"] = 300
```

# Preparation

```python
# Load the .csv file
df = pd.read_csv('/kaggle/input/wind-lidar-wind-profiles-measured-at-north-sea/ZephIR_windlidar_BSA_1s_201911210000_201911212350_v1.CSV')
```

```python
%%capture head
# Print the first 5 rows of the dataset
print(df.head())
```

# Display

```python
# Select a variable to plot
wind_speed = df['38m is a fixed reference measurement'][1:].values[:-1]
print(wind_speed)


fig, ax = plt.subplots()


# Plot the variable
ax.plot(wind_speed)


new_labels = []

for i in ax.yaxis.get_ticklabels():
    label = i.get_text()
    if len(label) > 10:
        new_labels.append(label[0:10])
    else:    
        new_labels.append(label)


ax.yaxis.set_ticklabels(new_labels)


plt.rcParams.update({'figure.autolayout': True})

plt.show()
```
