# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt



plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['figure.dpi'] = 300
```

# Load the dataset

```python
df = pd.read_csv('/kaggle/input/zambia-health-center-facilities/ZambiaHealthCenters.csv')
```

# Explore the data

```python
%%capture head
print(df.head())
```

```python
%%capture desc
print(df.describe())
```

# Rename columns

```python
df = df.rename(columns = {"ID_,C,16": "ID", "NAME1_,C,64": "Name", "NAME2_,C,64": "Type"})
```

# Visualize the distribution of facility types

```python
df['Type'].value_counts().plot(kind = 'bar')
plt.xlabel("Type")
plt.ylabel("Number of Facilities")
plt.show()
```

# Check for missing values

```python
print(df.isnull().sum())
```

# Drop rows with missing values

```python
df = df.dropna()
```

```python
# Visualize the distribution of locations
plt.figure(figsize = (96, 18), dpi = 300)
df['Name'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Save the cleaned dataset
df.to_csv('cleaned-zambia-health-center-facilities.csv', index=False)
```
