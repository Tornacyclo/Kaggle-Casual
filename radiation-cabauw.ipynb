# Import necessary libraries

```python
import xarray as xr
import matplotlib.pyplot as plt



plt.rcParams["figure.figsize"] = [32, 9]
plt.rcParams["figure.dpi"] = 300
```

# Load the .nc file

The xarray package is built on the NetCDF file format. 
NetCDF (Network Common Data Form) is one of the most popular data formats in climate and weather sciences.
NetCDF files usually contain many variables, dimensions, and attributes. 
Xarrayâ€™s main data structure is the Dataset. 
A Dataset is a multidimensional array, with associated coordinates and attributes.
xarray supports reading datasets from a variety of formats.

```python
ds = xr.open_dataset('/kaggle/input/radiation-validated-surface-long/cesar_surface_radiation_lb1_t10_v1.0_200005.nc')
```

# Print the available variables in the dataset

```python
%%capture data
print(ds.data_vars)
```

# Select a variable to plot

```python
radiation = ds['SWD']
```

# Display

```python
# Plot the variable
radiation.plot()
plt.title('Surface Radiation Data')


plt.show()
```

```python
%%capture attrs
# print the metadata information of the netcdf file
print(ds.attrs)


# Close the dataset
ds.close()
```
