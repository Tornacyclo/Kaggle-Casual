# Import necessary libraries

```python
import xarray as xr
import matplotlib.pyplot as plt



plt.rcParams["figure.figsize"] = [16, 9]
plt.rcParams["figure.dpi"] = 300
```

# Preparation

```python
%%capture data_vars
# Load the .nc file
ds = xr.open_dataset('/kaggle/input/wins50-wind-at-10600-meter-height/WINS50_43h21_fERA5_WFP_ptA_NETHERLANDS.NL_ix001_iy003_2019010100-2022010100.nc')


# Print the available variables in the dataset
print("Variables in the dataset:")
print(ds.data_vars)
```

# Display

```python
# Select a variable to plot
wind_speed = ds['wspeed']
```

```python
# Plot the variable
wind_speed.plot()


plt.title("Wind Speed")

plt.show()
```

```python
%%capture attrs
# Print the metadata information of the netcdf file
print(ds.attrs)


# Close the dataset
ds.close()
```

This notebook uses the xarray library to open and explore the .nc file. It first prints the available variables in the dataset, allowing you to select a specific variable to plot. In this case, I've selected the variable 'wind_speed' which stands for wind speed. The notebook then plots the selected variable and shows the plot. Next, it prints the metadata information of the netcdf file, and finally closes the dataset.
You may want to try different variables and do some analysis on the time series. Also, it is important to read the documentation provided in the link for better understanding of the dataset variables and attributes.
