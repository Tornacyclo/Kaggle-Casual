# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



plt.rcParams["figure.figsize"] = [16, 9]
plt.rcParams["figure.dpi"] = 300
```

This code will load the data from the .csv file into a Pandas dataframe, df, and provide a summary of the data using the describe() method.

Next, the code will plot the distribution of the number of persons in the images and the gender of the person in the images.

Finally, the code will get the count of images with one person and the gender distribution of the person in those images.

```python
%%capture describe
# Load the data from the .csv file
df = pd.read_csv('/kaggle/input/wikipedia-image-categorization/Wiki-gender-image-categorization-DFE.csv')

# Get a summary of the data
print(df.describe())
```

```python
# Plot the distribution of the number of persons in the images
df['no_of_persons'].value_counts().plot(kind = 'bar')
plt.xlabel('Number of Persons')
plt.ylabel('Count')
plt.title('Distribution of Number of Persons in Images')

plt.show()
```

```python
# Plot the distribution of the gender of the person in the images
df['person_singular_dom'].value_counts().plot(kind = 'bar')
plt.xlabel('Gender of Person')
plt.ylabel('Count')
plt.title('Distribution of Gender of Person in Images')

plt.show()
```

```python
# Get the count of images with one person and the gender of the person
one_person_df = df[df['no_of_persons'] == 'One Person']
print('Count of images with one person:', one_person_df.shape[0])
print('Gender distribution of person in images with one person:')
print(one_person_df['person_singular_dom'].value_counts())
```
