```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt



plt.rcParams["figure.figsize"] = [32, 18]
plt.rcParams["figure.dpi"] = 300
```

# Preparation

```python
%%capture
!pip install xlrd
```

```python
# Load the .xls file
df = pd.read_excel('/kaggle/input/meteo-data-observations-from-decima/decima_data/1825_TEMP_Siebold_lists.xls', skiprows = 2)#, header = [0,1], index_col = [0])


# Drop unwanted columns
# df = df.drop(df.columns[3:], axis=1)


# Print the first 5 rows of the dataset
print(df.head())
```

```python
# Rename the columns

# df.columns = df.columns.map('_'.join)

df = df.rename(columns={df.columns[1]:"Thermometer_morg", df.columns[2]:"Thermometer_mit", df.columns[3]:"Thermometer_abent"})
df = df.rename(columns={df.columns[4]:"Hygrometer_morg", df.columns[5]:"Hygrometer_mit", df.columns[6]:"Hygrometer_abent"})
```

# Display

```python
#Plot the temperature
df[["Thermometer_morg", "Thermometer_mit", "Thermometer_abent"]].plot()
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
#Plot the humidity
df[["Hygrometer_morg", "Hygrometer_mit", "Hygrometer_abent"]].plot()
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
%%capture describe
# Print the metadata information of the dataframe
print(df.describe())
```

This notebook uses the pandas library to open and explore the .xls file. It first prints the first 5 rows of the dataset, allowing you to have an overview of the data.
