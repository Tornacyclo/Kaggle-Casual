# Group by

```python
# Group the data by 'Country of publication' and count the number of entries for each country
print(df.groupby(['Country of publication']).size())
```

```python
# Group the data by 'Subject/genre terms' and count the number of entries for each subject/genre
print(df.groupby(['Subject/genre terms']).size())
```

```python
# Group the data by 'Publisher' and count the number of entries for each publisher
print(df.groupby(['Publisher']).size())
```

```python
# Group the data by 'Composer' and count the number of entries for each composer
print(df.groupby(['Composer']).size())
```

```python
# Group the data by 'Series title' and count the number of entries for each series
print(df.groupby(['Series title']).size())
```

```python
# Group the data by 'Notes' and count the number of entries for each note
print(df.groupby(['Notes']).size())
```

# Graph

```python
plt.figure(figsize = (64, 9), dpi = 300)



country_count = df.groupby(['Country of publication']).size()


# Create a bar graph of the number of entries for each country
country_count.plot(kind = 'bar')


# Add a title and x and y labels to the graph
plt.title('Number of Entries per Country of Publication')
plt.xlabel('Country of Publication')
plt.ylabel('Number of Entries')

plt.show()
```
