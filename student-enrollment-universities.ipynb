# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



plt.rcParams["figure.figsize"] = [16, 9]
plt.rcParams["figure.dpi"] = 300
```

# Preparation

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/student-enrollment-in-public-universities/student-enrollment-in-public-universities-2007-08-to-2012-13-csv-1.csv', index_col = 'Fields')
print(df.columns)
# df2 = df.drop(df.columns[0],axis = 1)


# Transpose the DataFrame
df_transposed = df.transpose()
# df_transposed.drop(index = df_transposed.index[0], axis = 0, inplace = True)
```

```python
# Get the first 5 rows of the dataset
print(df_transposed.head())
```

```python
%%capture describe
# Get the summary statistics of the dataset
print(df_transposed.describe())
```

# Plot the enrollment of undergraduates, postgraduates and other students

```python
plt.plot(df_transposed['Undergraduates'], label = 'Undergraduates')
plt.plot(df_transposed['Postgraduates'], label = 'Postgraduates')
plt.plot(df_transposed['Other, i.e. Diploma, etc.'], label = 'Other')


plt.legend()

plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Enrollment of Undergraduates, Postgraduates and Other students in Public Universities')

plt.show()
```

# Plot the enrollment of each student group as a bar chart

```python
df_transposed.plot(kind = 'bar', stacked = True)


plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Enrollment of Undergraduates, Postgraduates and Other students in Public Universities')

plt.show()
```

# Plot the enrollment of each student group as a heatmap

```python
sns.heatmap(df_transposed.corr(), annot = True)


plt.title('Correlation between Enrollment of Undergraduates, Postgraduates and Other students in Public Universities')

plt.show()
```

# Plot the distribution of the enrollment of each student group

```python
sns.displot(df_transposed, kind = 'hist')


plt.xlabel('Enrollment')
plt.ylabel('Density')
plt.title('Distribution of Enrollment of Undergraduates, Postgraduates and Other students in Public Universities')

plt.show()
```
